# Project: Iris Data Set

The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. One class is linearly separable from the other 2; the latter are NOT linearly separable from each other. 

## Getting Started

To begin working with the Iris data set, we'll first need to import the functionality we need, and load our data into a pandas DataFrame.
Run the code cell below to load our data and display the first few entries (sepals and petals dimensions of flowers) for examination using the .head() function.

```python
# import visuals as vs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from IPython.display import display # Allows the use of display() for DataFrames

# Import supplementary visualizations code visuals.py


# Pretty display for notebooks
%matplotlib inline

# Load the dataset
in_file = 'irisdataset.csv'
full_data = pd.read_csv(in_file)

# Print the first few entries of the RMS Titanic data
display(full_data.head())
```

From a sample of the Iris data, we can see the various features present for each flower:

##### sepal length :  Gives the sepal length of a flower 

##### sepal width : Gives the sepal width of a flower

##### petal length : Gives the petal length of a flower 

##### petal width : Gives the petal width of a flower

##### class : Gives the name of the flower

We need to convert the string representation of flower category into integers such as 0,1,2
since various data visualzation techniques uses numerical values but not strings 

Even though we have strings it will be converted into some floating values at the time of representations

```python
full_data['class'] = full_data['class'].map({'Iris-setosa': 0,'Iris-versicolor': 1,'Iris-virginica': 2})

display(full_data.head())
```

Since we're interested in the outcome of flower category based on the flower dimensions, we can remove the Class feature from this dataset and store it as its own separate variable results. We will use these results as our prediction targets.

Run the code cell below to remove Class as a feature of the dataset and store it in results.

While dropping the class column, we are appending 'axis = 1' this will indicate that we are dealing with columns 

```python
result = full_data['class']

data = full_data.drop('class' , axis = 1)

display(data.head())
```
